##Problem 1: Mobile Recharge Validation System

```python
# Function to check recharge validity
def check_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Amount must be at least ₹50")
        return False

    if amount not in valid_plans:
        print("Invalid plan")
        return False

    return True


# Retry until valid
while True:
    amount = int(input("Enter recharge amount: "))

    if check_recharge(amount):
        print("Recharge successful!")
        break
    else:
        print("Try again\n")
```

##Problem 2: Inventory Reorder Alert System

```python
# Function to check inventory
def check_inventory(products):
    for item, stock in products.items():
        if stock < 15:
            print(item, "- Reorder Alert")
        else:
            print(item, "- Stock OK")


# Sample dictionary
products = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18
}

# Call function
check_inventory(products)
```

## Problem 3: Student Result Processing System

```python
# Function to process student result
def check_result(marks):
    total = 0

    # Calculate total using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Example list
student_marks = [60, 45, 70, 55]

# Call function
result = check_result(student_marks)
print("Result:", result)
```

##Problem 4: Cab Fare Estimator with Retry Option

```python
# Function to calculate fare
def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12

    fare = base_fare + (distance * per_km)

    if peak == "yes":
        fare = fare * 1.25   # 25% extra

    return fare


# Retry option using while loop
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

##Problem 5: Employee Attendance Eligibility Checker

```python
# Function to check attendance eligibility
def check_attendance(attendance_list):

    present_days = 0

    # Count present days
    for day in attendance_list:
        if day == "P":
            present_days += 1

    # Calculate percentage
    percentage = (present_days / len(attendance_list)) * 100

    # Check eligibility
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


# Example usage
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]
result = check_attendance(attendance)

print("Result:", result)
```

##Problem 6: Password Strength Checker

```python
# Function to check password strength
def check_password(password):

    has_digit = False
    has_special = False

    special_chars = "@#$"

    # Check each character
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    # Apply rules
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# Example usage
pwd = input("Enter password: ")
result = check_password(pwd)
print(result)
```
