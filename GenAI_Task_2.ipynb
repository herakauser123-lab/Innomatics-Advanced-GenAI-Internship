###Problem 1: Unique Words in a Sentence

```python
def unique_words(sentence):
  #convert into lowercase
  sentence = sentence.lower()

  #split into words
  words=sentence.split()

  #remove duplicates
  unique = set(words)

  return unique, len(unique)
sentence = input("Enter Sentence: ")

unique_set, count = unique_words(sentence)

print("Unique words count: ", count)
print("Unique words: ", unique_set)
```

###Problem 2: Highest Salary from Employee Data

```python
def find_highest_salary(employees):
  # find employee with highest salary
  highest_employee = max(employees, key=employees.get)
  highest_salary = employees[highest_employee]

  return highest_employee,highest_salary

employees = {
    "Ravi" : 75000,
    "Anita" : 68000,
    "Kiran" : 72000
}

name, salary = find_highest_salary(employees)

print("Highest Salary: ", name, salary)
```

###Problem 3: Find Maximum and Minimum Values

```python
def find_max_min(numbers):
    maximum = numbers[0]
    minimum = numbers[0]

    for num in numbers:
        if num > maximum:
            maximum = num
        if num < minimum:
            minimum = num

    return maximum, minimum


numbers = [45, 22, 89, 10, 66]

max_val, min_val = find_max_min(numbers)

print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

###Problem 4: Count Products Above a Price Threshold

```python
def count_above_threshold(prices, threshold):
    count = 0

    for price in prices:
        if price > threshold:
            count += 1

    return count


prices = [450, 1200, 899, 1500, 300]

result = count_above_threshold(prices, 1000)

print("Prices:", prices)
print("Products above 1000:", result)
```

###Problem 5: Calculate Attendance Percentage

```python
def calculate_attendance(attendance):
    p_count = 0

    for status in attendance:
        if status == "P":
            p_count += 1

    percentage = (p_count / len(attendance)) * 100

    return percentage


attendance = ["P", "P", "A", "P", "P"]

result = calculate_attendance(attendance)

print("Attendance List:", attendance)
print("Attendance Percentage:", result)
```

###Problem 7: Count Character Frequency

```python
def count_character_frequency(text):
    frequency = {}

    for char in text:
        if char in frequency:
            frequency[char] += 1
        else:
            frequency[char] = 1

    return frequency


text = "pythonp"

result = count_character_frequency(text)

print("Text:", text)
print("Character Frequency:", result)
```

```python
def count_character_frequency(text):
    text = text.lower().replace(" ", "")
    frequency = {}

    for char in text:
        if char in frequency:
            frequency[char] += 1
        else:
            frequency[char] = 1

    return frequency


text = input("enter your text here: ")

result = count_character_frequency(text)

print("Text:", text)
print("Character Frequency:", result)
```

###Problem 10: Calculate Average Marks

```python
def calculate_average(marks):
    total = sum(marks)
    count = len(marks)
    average = total / count
    return average


marks = [65, 70, 80, 75]

avg = calculate_average(marks)

print("Marks:", marks)
print("Average Marks:", avg)
```
