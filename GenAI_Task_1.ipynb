```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

#Calculate pass and fail counts
for mark in marks:
    if mark >= 50:
        print(f"{mark} = Pass")
        pass_count += 1
    else:
        print(f"{mark} = Fail")
        fail_count += 1

#Print counts
print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
def clean_names(names):
    cleaned_list = []

#Clean data
    for name in names:
        cleaned_name = name.strip().lower()
        cleaned_list.append(cleaned_name)

    return cleaned_list


names = ["Alice ", " bob", "CHARLIE"]

cleaned_names = clean_names(names)

#Print cleaned data
print("Cleaned Names:", cleaned_names)
```
