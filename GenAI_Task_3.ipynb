###Problem Statement 1: Employee Performance Bonus Eligibility

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find highest score
highest_score = max(employees.values())

# Step 2: Find all employees with highest score
top_performers = []

for name in employees:
    if employees[name] == highest_score:
        top_performers.append(name)

# Step 3: Display result
print("Top Performers Eligible for Bonus:", top_performers)
print("Score:" highest_score)
```

###Problem Statement 2: Search Query Keyword Analysis

```python
sentence = "Buy mobile phone buy phone online"

# Step 1: convert to lowercase
sentence = sentence.lower()

# Step 2: split into words
words = sentence.split()

# Step 3: count frequency
word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Step 4: display words searched more than once
result = {}

for word in word_count:
    if word_count[word] > 1:
        result[word] = word_count[word]

print(result)
```

###Problem Statement 3: Sensor Data Validation

```python
# List of sensor readings where index represents hour
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Empty list to store valid (even) readings
valid_readings = []

# Loop through list with index and value using enumerate()
for index, value in enumerate(sensor_readings):

    # Check if the reading is even (valid reading)
    if value % 2 == 0:

        # Store the result as (hour_index, reading_value)
        valid_readings.append((index, value))

# Display the final list of valid readings
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

###Problem Statement 4: Email Domain Usage Analysis

```python
# List of employee email IDs
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Dictionary to store domain counts
domain_count = {}

# Loop through each email
for email in emails:

    # Extract domain after '@'
    domain = email.split("@")[1]

    # Count occurrences of each domain
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Total number of emails
total_emails = len(emails)

# Calculate and print percentage usage
for domain in domain_count:
    percentage = (domain_count[domain] / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

###Problem Statement 5: Sales Spike Detection

```python
#list of daily sales
sales = [1200, 1500, 900, 2200, 1400, 3000]

#calculate average sales
average_sales = sum(sales)/len(sales)

#calculate 30% above
for i in range(len(sales)):
  if sales[i] > average_sales * 1.30:
    #print 30% above sales
    print("Day", i+1 , ":" , sales[i])
```

```python
# Given expected output is day 4: 2200 , day 6: 3000
# Average sales = 1700
# 30% above average = 1700 Ã— 1.30 = 2210
# Day 4 sales = 2200, which is less than 2210, so it does not satisfy the condition of being more than 30% above average.
# Only Day 6 (3000) is above 2210.
# Therefore, the expected output is incorrect because Day 4 should not be included.
```

###Problem Statement 6: Duplicate User ID Detection

```python
# List of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to store frequency of each user ID
count_dict = {}

# Count occurrences
for user in user_ids:
    if user in count_dict:
        count_dict[user] += 1
    else:
        count_dict[user] = 1

# Print only duplicates
for user, count in count_dict.items():
    if count > 1:
        print(user, "->", count, "times")
```
