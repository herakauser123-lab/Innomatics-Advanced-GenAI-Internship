##Problem 1: Social Media – Post Engagement Analyzer

```python
# Function to analyze post engagement
def analyze_engagement(likes_list):

    total_likes = 0

    # Calculate total likes using loop
    for likes in likes_list:
        total_likes += likes

    # Check status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Example usage
likes = [200, 300, 150, 400]   # Total = 1050
analyze_engagement(likes)
```

##Problem 3: Agriculture – Rainfall Adequacy Checker

```python
# Function to check rainfall adequacy
def check_rainfall(rainfall_list):

    total = 0
    required_level = 70
    # Calculate total rainfall using loop
    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)


# Example usage
rainfall_data = [70, 75, 68, 80, 67]   # Average = 72
check_rainfall(rainfall_data)
```

##Problem 4: Social Media – Duplicate Account Detection

```python
# Function to check duplicate usernames
def check_duplicates(usernames):

    # Convert list to set
    unique_users = set(usernames)

    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example usage
users = ["hera", "kiran", "anita", "hera"]
check_duplicates(users)
```

##Problem 6: Agriculture – Premium Crop Price Filter

```python
# Function to filter premium crop prices
def filter_premium_crops(prices):

    premium_crops = []

    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)


# Example usage
crop_prices = [1500, 2500, 1800, 3200, 2000]
filter_premium_crops(crop_prices)
```

##Problem 9: E-Learning – Student Attendance Eligibility System

```python
# Function to calculate attendance and check eligibility
def check_exam_eligibility(attendance_list):

    total_days = len(attendance_list)
    present_days = 0

    # Count present days
    for day in attendance_list:
        if day == "P":
            present_days += 1

    # Calculate percentage
    attendance_percentage = (present_days / total_days) * 100

    # Check eligibility
    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", attendance_percentage)
    print("Exam Eligibility:", status)


# Example usage
attendance = ["P", "P", "A", "P", "P"]   # 4 out of 5 = 80%
check_exam_eligibility(attendance)
```

##Problem 10: Smart Electricity Bill Analyzer

```python
# Function to calculate electricity bill and usage level
def calculate_bill(units):

    # Bill calculation using slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage


# Example usage
total_bill, status = calculate_bill(250)

print("Total Bill:", total_bill)
print("Usage Level:", status)
```
